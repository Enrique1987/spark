### Save Excel with Image

```
from shutil import copyfile
import xlsxwriter

output_path = "/dbfs/my_path/"
file_name = "file_name_" + datetime.today().strftime('%Y_%m_%d')
temp_file = '/tmp/PRODUCT2.xlsx'

workbook = xlsxwriter.Workbook(temp_file)
worksheet = workbook.add_worksheet("my_sheet_name")
worksheet.set_column('A:A', 30)
bold = workbook.add_format({'bold': True})
# Insert an image.
worksheet.write('A2', 'Image Title', bold)
worksheet.insert_image('B2', '/dbfs/' + path_bild)

worksheet.write('A31', 'Title DataSet', bold)
worksheet.write('B32', 'column1', bold)
worksheet.write('C32', 'column2', bold)

for index, row in df[:10].iterrows(): # if our df had two columns 
    excel_row_word  = 'B' + str(index + 33)
    excel_row_count  = 'C' + str(index + 33) 
    worksheet.write(excel_row_word, row['colum1'])
    worksheet.write(excel_row_count, row['column2']) 

workbook.close()
copyfile(temp_file, output_path + file_name)
```
